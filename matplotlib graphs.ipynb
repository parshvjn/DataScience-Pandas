```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Bar Plot

```python
x = ['a','b','c','d']
y = [1,5,3,2]
```

```python
plt.figure(figsize=(10,5))
plt.grid(lw=1)
plt.xlabel('label')
plt.ylabel('Y-label')
plt.bar(x,y,width=0.4,animated=False)
```

#### Merged bar graph

```python
x1,x2 = ['a','b','c','d'],['e','eb','ec','ed']
y1,y2 = [1,5,3,2],[13,3,2,1]
```

```python
plt.grid(lw=1)
plt.bar(x1,y1,label='Alpha')
plt.bar(x2,y2,label='beta')
plt.legend()
```

### Double bargarph

```python
plt.figure(figsize=(15,5))
plt.grid()
X = ['group A','group B','group C','group D' ]
x1,x2 = [10,15,30,12],[13,30,20,10]
x_range = np.arange(len(x1))

plt.bar(x_range-0.2,x1,width=0.4,color = 'b',label = 'Alpha')
plt.bar(x_range+0.2,x2,width=0.4,color = 'm',label = 'Beta')
plt.xticks(x_range,X)
plt.legend()
plt.show()
```

```python
high_time_list = list(high_dict.keys())
```

```python
high_value_list = list(high_dict.values())
```

```python
times = set(low_time + high_time)
```

```python
times = sorted(times)
```

```python
data = {}
for time in times:
    if time in low_time and time in high_time:
        data[time] = {"High": high_value[high_time.index(time)], "low": low_value[low_time.index(time)]}
    elif time in low_time:
        data[time] = {"low": low_value[low_time.index(time)]}
    else:
        data[time] = {"High": high_value[high_time.index(time)]}
```

```python
sin = np.linspace(-2*np.pi, 2*np.pi, 200)
```

```python
plt.plot(np.sin(sin), marker = "x", color = "red", label = "sine")
plt.plot(np.cos(sin), marker = "*", color = "green", label = "cos")
plt.grid()
plt.title('Sine and Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 1)
```
