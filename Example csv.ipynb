```python
# HW:
# Q1: Write a program to find the fibbonacci series.
# Q2: Write a program to find the prime numbers between 25 to 105
# Q3: Write a program to calculate the bmi (body mass index) of a person having height of 167 cm and weight of 30 lb
# also distinguish the bmi index with the defined standards. (formula on google)
# Ranges:
# BMI <17.0: thinness
# BMI <18.5: underweight
# BMI 18.5-24.9: normal weight
# BMI ≥25.0: overweight
# BMI ≥30.0: obesity.
# Q4: Write a program to find the factorial of 5,6, and 7.
# Q5: Write a program to find the number of combinations that can be formed with 5 players in a batting order
# without using built-in functions.
```

```python
#Q1:

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return (fibonacci(n-1) + fibonacci(n-2))
    
n = 20
for x in range(n):
    print(fibonacci(x), end = ' ~ ')
```

```python
#Q2:

n1 = 25
n2 = 105

for x in range(n1, n2+1):
    if x > 1:
        for i in range(2, x):
            if x%i == 0:
                break
        else:
            print(x)
```

```python
#Q3: 

def bmi(w, h):
    a = (w*0.453592) #lb to kg
    b = h/100
    return a / (b**2)

h = 167 #cm
w = 30 #lb

print("BMI:", bmi(w,h))
if bmi(w,h) < 17:
    print("Thinness")
elif bmi(w,h) < 18.5:
    print("Underwheight")
elif bmi(w,h) > 18.5 and bmi(w,h) < 24.9:
    print("Normal")
elif bmi(w,h) >= 25:
    print("Overwheight")
elif bmi(w,h) >= 30:
    print("Obesity")
```

```python
#Q4:

def factorial(x):
    if x == 0 or x == 1:
        return 1
    else:
        return x*factorial(x-1)
    
print(factorial(5))
print(factorial(6))
print(factorial(7))
```

```python
# Q5: 

def comb(n, r):
    if r<=n:
        return factorial(n)/factorial(n-r)*factorial(r)
print(comb(3, 2))
```
