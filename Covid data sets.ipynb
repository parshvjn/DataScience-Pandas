```python
# Find which country has max number of 1st vaccination dose.
k = 0
l = []
for x in data1["PERSONS_VACCINATED_1PLUS_DOSE"]:
    if x == max(data1["PERSONS_VACCINATED_1PLUS_DOSE"]):
        l.append(k)
    else:
        k+=1
for x in l:
    print(data1.iloc[x])
```

```python
# Find which country has min nunber of booster dose.
j = 0
l = []
for x in data1["PERSONS_BOOSTER_ADD_DOSE"]:
    if x == min(data1["PERSONS_BOOSTER_ADD_DOSE"]):
        l.append(j)
    else:
        j+=1
for x in l:
    print(data1.iloc[x]["COUNTRY"])
```

```python
grouped = data1.groupby(by = "WHO_REGION").sum()
```

```python
grouped.columns
```

```python
grouped["percentage_of_people_with_1_plus_dose"] = grouped.apply(lambda x : (x["PERSONS_VACCINATED_1PLUS_DOSE"]/x["TOTAL_VACCINATIONS"])*100, axis=1)
```

```python
grouped
```

```python
grouped["percentage_of_people_with_booster_dose"] = grouped.apply(lambda x : (x["PERSONS_BOOSTER_ADD_DOSE"]/x["TOTAL_VACCINATIONS"])*100, axis=1)
```
