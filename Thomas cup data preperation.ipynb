```python
def pattern_match(s):
    import re
    for i in eval(s):
        match = re.match("ThomasCup".lower(), i.lower())
#     match = re.match("ThomasCup".lower(), s)
    if match == None or match == []:
        return None
    else:
        return match.group()


df_copy["ThomasCupIsIn"] = df_copy.apply(lambda x : pattern_match(str(x["hashtags"])), axis = 1)

#Hw : Create a new column and use apply or lambda function, take the result and store it in the column
```
